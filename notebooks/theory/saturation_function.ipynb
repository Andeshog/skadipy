```python
import numpy as np
import matplotlib.pyplot as plt
# use latex
plt.rc('text', usetex=True)
```

```python
n = 100
x = np.linspace(-10.0, 10.0, n)

zeta = [0.0, 0.1, 0.5, 1, 2, 5, 10]


y = np.empty((n, len(zeta)))
for i, z in enumerate(zeta):
    y[:, i] = x / (np.abs(x) + z)

for i in range(len(zeta)):
    plt.figure(1)
    plt.plot(x, y[:, i], label=r'$\zeta = $' + str(zeta[i]))

plt.xlabel(r'$x_1$')
plt.ylabel(r'$y_1$')
plt.grid(True)
plt.legend()
```
