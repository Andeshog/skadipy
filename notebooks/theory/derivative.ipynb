```python
%reset -f

import sys;
sys.path.append('../')
sys.path.append('../../src/')

from pythonrc import *
import mccontrolpy as mc
import numpy as np
import scipy as sc
import matplotlib.pyplot as plt

import ipywidgets as widgets
```

```python
# Generate sin signal and add noise
N = 1000
t = np.linspace(0, 20, N)
x = np.sin(t)

dx = np.gradient(x)

epf = skadipy.toolbox.derivative.ExponentialSmoothing(r = 0.1)
dx_epf = np.zeros_like(x)
for i in range(len(x)):
    dx_epf[i] = epf(x[i])


# plot the signal
plt.figure()
plt.plot(t, x, 'k')
plt.plot(t, dx, 'r')
plt.plot(t, dx_epf, 'b')
plt.grid(True)
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.legend(['Signal', 'Numerical Derivative', 'Exponential Smoothing'])
plt.title('Signal')
plt.show()
```
